```python
from ultralytics import YOLO
```

```python
BASE_DIR = Path(__file__).parent.parent.parent.parent
```

```python
# === CONFIG ===
DATA_YAML = BASE_DIR / "data" / "yolo" / "wgisd.yaml"
PRETRAINED_MODEL = BASE_DIR / "models" / "yolov8s.pt"   # modelo peque√±o preentrenado, se puede cambiar a yolov8s/m/l/x
OUTPUT_DIR = BASE_DIR / "src" / "detection" / "training" / "training_results"
FOLDER_NAME = "training_1"
EPOCHS = 100
IMG_SIZE = 640
BATCH_SIZE = 8
```

```python
model = YOLO(PRETRAINED_MODEL)
```

```python
# ‚ö° Transfer Learning: congelar el backbone y entrenar solo el head
# Esto se hace con parameter freezing: freeze=number of layers to freeze
# Por defecto YOLOv8 usa freeze=10, que congela el backbone
# Puedes ajustar seg√∫n necesidad
model.train(
    data=DATA_YAML,
    epochs=EPOCHS,
    imgsz=IMG_SIZE,
    batch=BATCH_SIZE,
    freeze=[10],  # congela primeras 10 capas (backbone), entrena head
    project=OUTPUT_DIR,
    name=FOLDER_NAME,
    exist_ok=True
)
```

```python
# Despu√©s de entrenar puedes guardar tu modelo final
#model.export(format="torchscript")  # o torchscript, coreml, etc.

model.save("wgisd-grape-cluster-detection.pt")
```
